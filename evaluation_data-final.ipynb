```python


# Initialize BigQuery client
client = bigquery.Client()

# Define queries and their labels
# Define queries and their labels
ground_truth_queries = [
    {
        "prompt": "How many bookings were made by customers in the 25-34 age group?",
        "output": "There were 805 bookings made by customers in the 25-34 age group."
    },
    {
        "prompt": "Show the average ticket price for international flights grouped by airline",
        "output": {
            "content": "Here's the average international ticket price for each airline:",
            "table": [
                {
                    "Airline": "Alaska Airlines",
                    "Average Ticket Price": "$1645.12"
                },
                {
                    "Airline": "Delta Airlines",
                    "Average Ticket Price": "$1722.42"
                },
                {
                    "Airline": "Southwest Airlines",
                    "Average Ticket Price": "$1720.21"
                },
                {
                    "Airline": "American Airlines",
                    "Average Ticket Price": "$1714.81"
                },
                {
                    "Airline": "United Airlines",
                    "Average Ticket Price": "$1688.75"
                }
            ],
            "footer": "Delta Airlines and Southwest Airlines have the highest average ticket prices for international flights, while Alaska Airlines has the lowest."
        }
    },
    {
        "prompt": "How many bookings were made by customers aged 35 to 44 with Southwest Airlines??",
        "output": "There were 188 bookings made by customers aged 35 to 44 with Southwest Airlines." 
    },
    {
        "prompt": "What is the average ticket price for business class bookings?",
        "output": "The average ticket price for business class bookings is $1181.97." 
    },
    {
        "prompt": "How many customers booked flights with Delta Airlines in September 2024?",
        "output": "There were 170 customers who booked flights with Delta Airlines in September 2024." 
    },
    {
        "prompt": "Which age group spends the most on flight tickets?",
        "output": "The age group that spends the most on flight tickets is 25-34, with an average spend of $1165.73." 
    },
    {
        "prompt": "What percentage of bookings were made through the mobile app?",
        "output": "Mobile app bookings accounted for 32.56% of total bookings." 
    },
    {
        "prompt": "What is the total revenue generated by ticket sales by Delta Airlines?",
        "output": "The total revenue generated from ticket sales by Delta Airlines is $1,116,843.03." 
    },
    {
        "prompt": "Which states have the highest number of frequent flyers?",
        "output": {
            "content": "Here are the top 5 states with the most frequent flyers:",
            "table": [
                {
                    "State": "Illinois",
                    "Frequent Flyer Count": "72"
                },
                {
                    "State": "California",
                    "Frequent Flyer Count": "72"
                },
                {
                    "State": "Georgia",
                    "Frequent Flyer Count": "65"
                },
                {
                    "State": "Oregon",
                    "Frequent Flyer Count": "62"
                },
                {
                    "State": "New York",
                    "Frequent Flyer Count": "62"
                }
            ],
            "Observation": " Illinois and California have the highest number of frequent flyers."
        }
    },
    {
        "prompt": "How many frequent flyers booked flights to international destinations?",
        "output": "There are 470 frequent flyers who booked flights to international destinations." 
    }
]


# List to store evaluation dataset entries
evaluation_dataset = []
```

```python

# Execute each query and store the expected results
for query_entry in ground_truth_queries:
    query = query_entry["prompt"]
    prompt = query_entry["output"]
    
    try:
        # Run the query
        query_job = client.query(query)
        results = query_job.result()
        
        # Convert results to a list of dictionaries
        expected_output = str([dict(row) for row in results])

        # Append to the evaluation dataset
        evaluation_dataset.append({
            "prompt": query,  # Store the original SQL query
            "output": prompt,  # Add a descriptive label for context
            #"expected_output": expected_output,  # Store the query results
            
        })

    except Exception as e:
        print(f"Error running query for '{prompt}': {e}")
```

```python

# Save the dataset to a JSON file
with open("evaluation_dataset_sample.json", "w") as f:
    json.dump(ground_truth_queries, f, indent=4)

print("Evaluation dataset generated and saved to 'evaluation_dataset_sample.json'")
```
